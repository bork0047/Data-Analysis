# 1. Creating a single neuron
A neuron accepts the incoming inputs. Multiplies each input with associated weight. Sums these together, then adds the bias. 

In essence the weights tell the neuron what fraction of the input should be considered. You can think of it like a computation approach to informing the neuron as to how important the inputs are - the higher the weight the more influence that input value will have. 

These weights together with the bias are the adjustable parameters of the neuron. 

````python
inputs = [1, 2, 3]
weights = [0.2, 0.8, -0.5]
bias = 1
````

````python
output = (inputs[0]*weights[0] +
          inputs[1]*weights[1] +
          inputs[2]*weights[2] + bias)
````

### Calculate the outputs for the neuron layer in a for loop

````python
inputs = [2,3,2.5]
weights = [[0.2, 0.8, -0.5], 
 [0.5, -0.91, 0.26], 
 [ -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]


# Output of current layer
layer_outputs = []
# For each neuron
for neuron_weights, neuron_bias in zip(weights, biases):
    # Zeroed output of given neuron
    neuron_output = 0
    # For each input and weight to the neuron
    for n_input, weight in zip(inputs, neuron_weights):
        # Multiply this input by associated weight
        # and add to the neuron’s output variable
        neuron_output += n_input*weight
    # Add bias
    neuron_output += neuron_bias
    # Put neuron’s result to the layer’s output list
    layer_outputs.append(neuron_output)

print("layer_outputs are => {}".format(layer_outputs))
````

# 3. Basic Dense Layer class
Using the ideas from above, lets start creating the basic class for a neuron layer. For this we will also create some data and look at activation functions needed by each neuron layer. 

### Generate some Training Data
Rather than manually adding sample input data. Lets create a method to generate some non-linear data.

````python
#using a data generator from https://cs231n.github.io/neural-networks-case-study/

def spiral_data (N, D, K):
  X = np.zeros((N*K,D)) # data matrix (each row = single example)
  y = np.zeros(N*K, dtype='uint8') # class labels
  for j in range(K):
    ix = range(N*j,N*(j+1))
    r = np.linspace(0.0,1,N) # radius
    t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # theta
    X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
    y[ix] = j


  return X, y

N = 100 # number of points per class
D = 2 # dimensionality
K = 3 # number of classes
X_data, y_data = spiral_data(N, D, K)
# lets visualize the data:
plt.scatter(X_data[:, 0], X_data[:, 1], c=y_data, s=40, cmap=plt.cm.Spectral)
plt.show()
````

Now that we have some data generated lets create a class for our neuron layer. Since each neuron in our layer is connected to each of the inputs; we refer to it as a fully connected layer or in short a dense layer. 

````python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        np.random.seed(0)
        # weights initialised from a gaussian with mean 0 and variance 1
        # multiply with a small value just to ensure we start with small weight values
        self.weights = 0.1*  np.random.randn(n_inputs, n_neurons)
        #self.weights = np.random.randn(n_inputs, n_neurons) / np.sqrt(n_inputs) #xavier initialisation for deep nets
        #self.weights = np.random.randn(n_inputs, n_neurons) / np.sqrt(n_inputs/2) #he et al initialisation for with Relus deep nets
        self.biases = np.zeros((1, n_neurons))

       
    def forward(self, inputs):
        self.inputs = inputs #need this when backpropagating so maintain it in the layer class
        z = np.dot(inputs , self.weights) + self.biases # weights dot product
        self.outputs = z
````

Lets consider first the Relu. If the weighted sum plus bias value is zero or less than zero, then the Relu unit will output a zero; otherwise it will output the z value. 

The bias when higher has an impact of making the neuron fire earlier (i.e. whilst normally negative input values result in 0; it will change this such that activation can happen earlier with negative values ). Whilst negative weights can have an inverse effect on the normal Relu function. 

````python
inputs = [0, -2, -1, 3.3, -2.7, 11, 2.2, -100] # some arbitary data points

output = []

# This is is the impact that Relu has
for i in inputs:
    if i > 0:
        output.append(i)
    else:
        output.append(0)

print(output)
````

````python
class Activation_ReLU:
    def forward(self, x):
        return(np.maximum(0, x)) # without the if statement we acheive the same with np.maximum of function


# complete the code here based on the exercises below
class Activation_Sigmoid:
    def forward(self, x):
        return (1 / (1 + np.exp(-x)))


class Activation_Linear:
    def forward(self, x, m=1):
        return(x * m) # linear value multiplied by the m constant
````

# Improved Layer_Dense class with activation

````python
# improved verison of our Layer_dense class
# where the forward pass can now also directly do the 
# activation there
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        # weights initialised from a gaussian with mean 0 and variance 1
        # multiply with a small value just to ensure we start with small weight values
        self.weights = 0.1*  np.random.normal(0.0, pow(n_inputs, -0.5), (n_inputs, n_neurons))
        self.biases = np.zeros((1, n_neurons))

       
    def forward(self, inputs, act):
        self.inputs = inputs #need this when backpropagating so maintain it in the layer class
        z = np.dot(inputs , self.weights) + self.biases # weights dot product
        self.outputs = act.forward(z) # nonlinear activation output of the layer
````

### Helper functions to test the dense layer
The MSE is a measure of the quality of an estimator—it is always non-negative, and values closer to zero are better.
see for instance a detailed discussion [here](https://en.wikipedia.org/wiki/Mean_squared_error).

Accuracy is another metric for evaluating classification models. Informally, accuracy is the fraction of predictions our model got right. 

If you want to find out more about different types of metrics there is an overview [here](https://towardsdatascience.com/metrics-to-evaluate-your-machine-learning-algorithm-f10ba6e38234). 

````python
def mean_squared_error(predictions, targets):
    """
    Calculates mean squared error of a model's predictions.
    """
    N=targets.size
    mse = ((targets - predictions) **2).sum() / (2*N)
    return mse


def accuracy(predictions, targets):
    """
    Calculates the accuracy of a model's predictions.
    """
    prediction_labels = np.argmax(predictions, axis=1)
    target_labels = np.argmax(targets, axis=1)
    predictions_correct = (prediction_labels == target_labels.round())
    accuracy = predictions_correct.mean()
    return accuracy
````
